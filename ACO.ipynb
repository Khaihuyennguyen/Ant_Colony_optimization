```python
# Create a Cell class
class Cell:
  def __init__(self, coordinates, pheromone = 0):
    self.coordinates = coordinates
    self.pheromone = pheromone
    self.connected_cells = []
  def add_connected_cells(self, cell):
    if cell not in self.connected_cells:
      self.connected_cells.append(cell)
  def set_pheromone(self, pheromone):
    self.pheromone = pheromone
  
```

```python
# Create Data Cells:
# Extract the cells that are not occupied by walls in to the dictinalrry
# The keys will be the coordinates 
# The values would be the the corresponding Cell objects

cells = {(i,j) : Cell((i,j)) for i in range(len(dataList)) for j in range(len(dataList)) if dataList[i][j] != '0'}
```
